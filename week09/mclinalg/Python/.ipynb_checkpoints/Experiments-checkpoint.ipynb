```python
def projection(xx,axis=0):
    res = []
    for x in xx: res.append(x[axis])
    return res
```

```python

plt.scatter(projection(xx,0), projection(xx,1), alpha=0.5)
```

```python

def plot_seq(seq, dim1, dim2, n):
    xx = [seq() for i in xrange(n)]
    plt.scatter(projection(xx,dim1), projection(xx,dim2), alpha=0.5)
    plt.title(seq.rng_type() + '; s=%d' % (s,))
    plt.xlabel('Dimention %d' % (dim1,))
    plt.ylabel('Dimention %d' % (dim2,))
    plt.savefig(seq.rng_type()+'dims%dand%d' % (dim1,dim2),dpi=100)
    plt.close()
    
```

```python
s = 40
seq_length=2**16
max_scrambled_digits=15

numseq = [Faur(s,seq_length,max_scrambled_digits,iflag) for iflag in (0,1,2,3)]

seq = numseq[0]
dim1 = 33
dim2 = 37
n = 200
for seq in numseq: plot_seq(seq,dim1,dim2,n)
#xx = [seq() for i in xrange(n)]
#plt.scatter(projection(xx,dim1), projection(xx,dim2), alpha=0.5)
#plt.title(seq.rng_type() + '; s=%d' % (s,))
#plt.xlabel('Dimention %d' % (dim1,))
#plt.ylabel('Dimention %d' % (dim2,))
#plt.savefig(seq.rng_type()+'dims%dand%d' % (dim1,dim2),dpi=100)
```
